```python
import matplotlib.pyplot as plt

# Datos
equipos = ["Equipo A", "Equipo B"]
goles = [3, 1]

# Gráfico
plt.figure()
plt.bar(equipos, goles)
plt.title("Goles en un partido de fútbol")
plt.xlabel("Equipos")
plt.ylabel("Goles")
plt.show()
```
