```python
import pandas as pd
import matplotlib.pyplot as plt
# Cargar dataset
df = pd.read_csv("salarios_mexico.csv")

# Ver las primeras filas
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure()
plt.hist(df["salario_mensual_mxn"], bins=20, color="skyblue", edgecolor="black")

plt.title("Distribución del salario mensual en México")
plt.xlabel("Salario mensual (MXN)")
plt.ylabel("Frecuencia")

plt.savefig("histograma_salarios.png")
plt.show()
```

```python
plt.figure()
plt.scatter(df["id_persona"], df["salario_mensual_mxn"], alpha=0.5)

plt.title("Salario mensual por persona")
plt.xlabel("ID de persona")
plt.ylabel("Salario mensual (MXN)")

plt.savefig("dispersion_id_salario.png")
plt.show()
```

```python
plt.figure()
plt.boxplot(df["salario_mensual_mxn"])

plt.title("Boxplot del salario mensual")
plt.ylabel("Salario mensual (MXN)")

plt.savefig("boxplot_salarios.png")
plt.show()
```
