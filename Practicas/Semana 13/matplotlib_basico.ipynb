```python
import matplotlib.pyplot as plt

# Datos
equipos = ["Equipo A", "Equipo B"]
goles = [3, 1]
# Gráfico
plt.figure()
plt.bar(equipos, goles)
plt.title("Goles en un partido de fútbol")
plt.xlabel("Equipos")
plt.ylabel("Goles")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos
tiendas = ["Coppel", "Elektra", "Suburbia", "Liverpool"]
articulos_vendidos = [120, 90, 150, 70]

# Gráfico
plt.figure()
plt.bar(tiendas, articulos_vendidos)
plt.title("Artículos vendidos por tienda")
plt.xlabel("Tienda")
plt.ylabel("Cantidad de artículos")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos
dias = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes"]
usuarios = [200, 240, 180, 300, 280]

# Gráfico
plt.figure()
plt.plot(dias, usuarios, marker='o')
plt.title("Usuarios que inician sesión en la página")
plt.xlabel("Día")
plt.ylabel("Número de usuarios")
plt.show()
```
