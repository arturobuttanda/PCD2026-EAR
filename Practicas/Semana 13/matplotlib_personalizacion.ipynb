```python
import matplotlib.pyplot as plt

# Datos
meses = ["Ene", "Feb", "Mar", "Abr", "May"]
ventas_2024 = [120, 150, 170, 160, 180]
ventas_2025 = [130, 160, 180, 175, 200]

plt.figure()
plt.plot(meses, ventas_2024, color="blue", linestyle="--", marker="o", label="Ventas 2024")
plt.plot(meses, ventas_2025, color="green", linestyle="-", marker="s", label="Ventas 2025")

plt.title("Comparaci√≥n de ventas mensuales")
plt.xlabel("Mes")
plt.ylabel("Ventas")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos
productos = ["Producto A", "Producto B", "Producto C"]
tienda_1 = [40, 55, 30]
tienda_2 = [50, 45, 35]

plt.figure()
plt.bar(productos, tienda_1, color="orange", label="Tienda 1")
plt.bar(productos, tienda_2, color="purple", alpha=0.7, label="Tienda 2")

plt.title("Art√≠culos vendidos por producto")
plt.xlabel("Producto")
plt.ylabel("Cantidad")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos
dias = ["Lun", "Mar", "Mi√©", "Jue", "Vie"]
usuarios_web = [200, 230, 210, 250, 300]
usuarios_app = [180, 220, 200, 240, 280]

fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Subplot 1
axs[0].plot(dias, usuarios_web, color="red", marker="o", linestyle="-")
axs[0].set_title("Usuarios Web")
axs[0].set_xlabel("D√≠a")
axs[0].set_ylabel("Usuarios")

# Subplot 2
axs[1].plot(dias, usuarios_app, color="blue", marker="^", linestyle="--")
axs[1].set_title("Usuarios App")
axs[1].set_xlabel("D√≠a")
axs[1].set_ylabel("Usuarios")

# T√≠tulo general
fig.suptitle("Usuarios que inician sesi√≥n por plataforma")

plt.tight_layout()
plt.show()
```
