```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Estilo general
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("salarios_mexico.csv")
df.head()
```

```python
sns.set_style("darkgrid")

plt.figure()
sns.histplot(df["salario_mensual_mxn"], bins=30, kde=True)

plt.title("Distribución del salario mensual en México")
plt.xlabel("Salario mensual (MXN)")
plt.ylabel("Frecuencia")

plt.show()
```

```python
sns.set_style("ticks")

plt.figure()
sns.countplot(data=df, x="sexo")

plt.title("Distribución de personas por sexo")
plt.xlabel("Sexo")
plt.ylabel("Cantidad de personas")

plt.show()
```

```python
sns.set_style("white")

plt.figure()
sns.barplot(
    data=df,
    x="sexo",
    y="salario_mensual_mxn",
    estimator="mean"
)

plt.title("Salario promedio mensual por sexo")
plt.xlabel("Sexo")
plt.ylabel("Salario promedio (MXN)")

plt.show()
```
