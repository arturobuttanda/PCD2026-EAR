
# EDA: Medidas de Tendencia Central y Dispersión

## Objetivo
Conectar la **estadística descriptiva** con el **Análisis Exploratorio de Datos (EDA)**
mediante el estudio de **múltiples datasets reales**, analizando:

- Media
- Mediana
- Moda
- Varianza
- Desviación estándar
- Rango

Cada análisis se refuerza con una **interpretación** y **visualizaciones sencillas**.


```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
```


## Dataset 1: Titanic

Dataset clásico con información de pasajeros del Titanic.
Analizaremos variables relacionadas con **edad** y **tarifa pagada**.


```python

titanic = sns.load_dataset("titanic")

cols_titanic = ["age", "fare"]
titanic_num = titanic[cols_titanic].dropna()

stats_titanic = pd.DataFrame({
    "Media": titanic_num.mean(),
    "Mediana": titanic_num.median(),
    "Moda": titanic_num.mode().iloc[0],
    "Varianza": titanic_num.var(),
    "Desviación estándar": titanic_num.std(),
    "Rango": titanic_num.max() - titanic_num.min()
})

stats_titanic
```


### Interpretación (Titanic)

- La **edad** presenta una dispersión considerable, indicando diversidad de pasajeros.
- La **tarifa** tiene una varianza alta, reflejando diferencias económicas marcadas.
- La media de la tarifa es mayor que la mediana, lo que sugiere **asimetría positiva**.


```python

plt.figure(figsize=(8,4))
sns.boxplot(data=titanic_num)
plt.title("Boxplot de Edad y Tarifa - Titanic")
plt.show()
```


## Dataset 2: Iris

Dataset clásico de botánica con mediciones físicas de flores.
Analizaremos dimensiones del **sépalos**.


```python

iris = sns.load_dataset("iris")

cols_iris = ["sepal_length", "sepal_width"]
iris_num = iris[cols_iris]

stats_iris = pd.DataFrame({
    "Media": iris_num.mean(),
    "Mediana": iris_num.median(),
    "Moda": iris_num.mode().iloc[0],
    "Varianza": iris_num.var(),
    "Desviación estándar": iris_num.std(),
    "Rango": iris_num.max() - iris_num.min()
})

stats_iris
```


### Interpretación (Iris)

- Las medidas del sépalo son relativamente **estables**, con baja dispersión.
- La cercanía entre media y mediana indica distribuciones **simétricas**.
- Es un dataset ideal para análisis comparativos y clasificación.


```python

plt.figure(figsize=(8,4))
sns.histplot(iris_num["sepal_length"], kde=True)
plt.title("Distribución del largo del sépalo - Iris")
plt.show()
```


## Dataset 3: Tips

Dataset de propinas en restaurantes.
Analizaremos el **total de la cuenta** y el **monto de la propina**.


```python

tips = sns.load_dataset("tips")

cols_tips = ["total_bill", "tip"]
tips_num = tips[cols_tips]

stats_tips = pd.DataFrame({
    "Media": tips_num.mean(),
    "Mediana": tips_num.median(),
    "Moda": tips_num.mode().iloc[0],
    "Varianza": tips_num.var(),
    "Desviación estándar": tips_num.std(),
    "Rango": tips_num.max() - tips_num.min()
})

stats_tips
```


### Interpretación (Tips)

- El **total de la cuenta** tiene mayor dispersión que la propina.
- La propina presenta valores más concentrados.
- Existen posibles **outliers**, visibles en el rango y la varianza.


```python

plt.figure(figsize=(8,4))
sns.scatterplot(data=tips_num, x="total_bill", y="tip")
plt.title("Relación entre total de la cuenta y propina")
plt.show()
```


## Conclusiones generales

- Las **medidas de tendencia central** permiten entender el valor típico de los datos.
- Las **medidas de dispersión** revelan variabilidad y posibles outliers.
- Al combinar estadística descriptiva con visualizaciones:
  - El EDA se vuelve más intuitivo
  - Se facilita la toma de decisiones
  - Se fortalecen proyectos de ciencia de datos
